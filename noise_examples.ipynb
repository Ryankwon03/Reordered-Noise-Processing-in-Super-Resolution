```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


image_paths = ['images/birds.png', 'images/img2.png', 'images/img3.png']
images = [Image.open(path) for path in image_paths]
```

Plot unmodified images

```python
plt.figure(figsize=(12, 6))
for i, img in enumerate(images):
    plt.subplot(3, 3, i+1)
    plt.imshow(img)
    plt.axis('off')
```

Salt and Pepper applied to the images

```python
snp = []
for img in images:
    output = np.array(img)
    probs = np.random.random(output.shape[:2])
    output[probs < 0.05] = 0
    output[probs > 1 - 0.05] = 255
    snp.append(Image.fromarray(output))

for i, img in enumerate(snp):
    plt.subplot(3, 3, i+4)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
```

Gaussian Noise applied to the images

```python
gaussian = []
for img in images:
    img_array = np.array(img)
    sigma = 0.1 ** 0.5
    gauss = np.random.normal(0, sigma, img_array.shape)
    noisy = img_array + gauss * 255
    gaussian.append(Image.fromarray(np.clip(noisy, 0, 255).astype(np.uint8)))

for i, img in enumerate(gaussian):
    plt.subplot(3, 3, i+7)
    plt.imshow(img, cmap='gray')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
