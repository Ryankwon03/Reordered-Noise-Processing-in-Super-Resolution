```python
from PIL import Image
import cv2
import scipy.ndimage 
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = Image.open("../../data/original/image1.jpg")
img2 = Image.open("../../data/original/image2.jpg")
img3 = Image.open("../../data/original/image3.jpg")
images = [img1, img2, img3]
for img in images:
    display(img)
    display(img.size)
```

```python
gaussian = []
for img in images:
    img_array = np.array(img)
    sigma = 0.1 ** 0.5
    gauss = np.random.normal(0, sigma, img_array.shape)
    noisy = img_array + gauss * 255
    gaussian.append(Image.fromarray(np.clip(noisy, 0, 255).astype(np.uint8)))

for img in gaussian:
    display(img)
    display(img.size)
```

```python
downsampled1 = []
for img in gaussian:
    downsampled1.append(img.resize((int(img.size[0]/2), int(img.size[1]/2)), Image.Resampling.LANCZOS))

for img in downsampled1:
    display(img)
    display(img.size)
```

```python
snp = []
for img in downsampled1:
    output = np.array(img)
    probs = np.random.random(output.shape[:2])
    output[probs < 0.05] = 0
    output[probs > 1 - 0.05] = 255
    snp.append(Image.fromarray(output))

for img in snp:
    display(img)
    display(img.size)
```

```python
downsampled2 = []
for img in snp:
    downsampled2.append(img.resize((int(img.size[0]/2), int(img.size[1]/2)), Image.Resampling.LANCZOS))

for img in downsampled2:
    display(img)
    display(img.size)
```
