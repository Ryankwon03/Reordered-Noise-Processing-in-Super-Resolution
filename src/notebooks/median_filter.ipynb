```python
from PIL import Image
import cv2
import scipy.ndimage 
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = Image.open("../../results/SR_noised_downsample/image1.jpg")
display(img)
print(img.size)
```

```python
def median_filter(image, kernel_size=3):
    """
    Apply a median filter to an image.

    Parameters:
    - image: Input image (in grayscale or color).
    - kernel_size: Size of the kernel (odd integer, e.g., 3, 5, 7).

    Returns:
    - filtered_image: The image after applying the median filter.
    """
    # Check if the kernel size is odd, if not, make it odd
    if kernel_size % 2 == 0:
        kernel_size += 1

    # Apply the median filter using OpenCV's built-in function
    filtered_image = cv2.medianBlur(image, kernel_size)

    return filtered_image
```

```python
img_np = np.array(img)
filtered_image = median_filter(img_np, 2)
filtered_img = Image.fromarray(filtered_image)
display(filtered_img)
print(filtered_img.size)
```
