```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
data = np.load('../../results/SR_downsample/downsample_3x256x256x3_1000.npz') #best --> iteration = 1000, timestep = 250
#data2 = np.load('../../results/SR_downsample/downsample_3x256x256x3_2000.npz') #for experiment
data_noised = np.load('../../results/SR_noised_downsample/input_3x256x256x3_1000.npz')
print("Keys in the .npz file:", data.files)
```

```python
import matplotlib.pyplot as plt

# Assuming data['arr_0'] and data2['arr_0'] are numpy arrays or lists of images
for i, (img1, img2) in enumerate(zip(data['arr_0'], data_noised['arr_0'])):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # Create a subplot with 1 row and 2 columns

    # Display the first image
    axes[0].imshow(img1, cmap='gray')  # Use 'cmap' if the image is grayscale
    axes[0].axis('off')  # Turn off axis for the first image

    # Display the second image
    axes[1].imshow(img2, cmap='gray')  # Use 'cmap' if the image is grayscale
    axes[1].axis('off')  # Turn off axis for the second image

    # Optional: Set titles for each subplot
    axes[0].set_title(f"Image {i} from data['arr_0']")
    axes[1].set_title(f"Image {i} from data_noised['arr_0']")
    
    image1 = Image.fromarray(img1)
    image2 = Image.fromarray(img2)
    #image1.save(f"../../results/SR_downsample/image{i+1}.jpg")
    image2.save(f"../../results/SR_noised_downsample/image{i+1}.jpg")

    plt.show()  # Show the subplot
```
